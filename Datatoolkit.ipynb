17.  Explain the difference between apply() and map() in Pandas?
    - Both apply() and map() are used to apply functions to elements in Pandas (specifically Series or DataFrames), but they have distinct use cases and behavior. Here's a breakdown of their key differences:

      General Usage
apply()
Applies a function along an axis (rows or columns) of a DataFrame or Series.
Can be used with both DataFrames and Series.
Allows you to apply a function on each row/column or entire Series.
More flexible, as you can apply custom functions or built-in ones across rows, columns, or even across axis (row-wise or column-wise).

      map()
Primarily used for Series (one-dimensional data).
Used to apply element-wise functions to each element of a Series.
Typically used with dictionaries, functions, or Series to map values in the Series to new values

   -  Behavior and Flexibility
apply()
Can work on both rows and columns in a DataFrame.
Supports lambda functions, custom functions, and built-in functions.
Works with functions that take more than one argument (e.g., functions that perform operations across multiple columns).
map()
Works only on Series, not DataFrames.
Only allows element-wise transformations (i.e., operates on each element independently).
You can map values to new values using a dictionary or Series or apply a function element-wise.

21. Why is NumPy‚Äôs array slicing faster than Python‚Äôs list slicing?
    - NumPy arrays are faster than Python lists for slicing due to several key optimizations at both the memory and implementation levels. Let's break down the main reasons why NumPy‚Äôs array slicing outperforms Python‚Äôs list slicing.

     Memory Layout and Contiguous Memory Blocks

     NumPy Arrays are stored in contiguous memory blocks. This means all elements are stored in one continuous block of memory, which allows for efficient indexing and slicing.
Python Lists are implemented as arrays of references to objects. Each element in a Python list is a reference to a separate object in memory, making access slower compared to NumPy arrays.
NumPy Memory Layout:
NumPy arrays are stored as C-style arrays where each element is placed consecutively in memory. This allows for cache optimization and quick access to elements during slicing.
Python List Layout:
A Python list is essentially a dynamic array of references to objects. This adds overhead because every time an element is accessed, the interpreter has to dereference the pointer and access the object in memory.
    
     Element-wise Operations in NumPy
When you slice a NumPy array, NumPy doesn‚Äôt actually create a new array. Instead, it returns a view of the original array (unless explicitly copied). This is known as view-based slicing, meaning no extra memory is allocated for the new sliced array.

     In contrast, Python lists always create a new list when sliced, which involves copying the elements into a new list.

      Vectorized Operations in NumPy

     NumPy‚Äôs vectorized operations allow batch processing of operations over the array, which are implemented in C and optimized for performance. This contrasts with Python‚Äôs lists, which rely on interpreted loops for operations
      Optimized C Implementation in NumPy
NumPy is implemented in C, a low-level programming language, which is close to the hardware and highly optimized for numerical operations. When slicing, NumPy makes direct references to the memory of the original array, minimizing the need for extra work during slicing.

     C-based Optimization:

     The underlying C implementation allows NumPy to minimize overhead and perform slicing operations with much greater efficiency compared to Python‚Äôs higher-level list operations.
In contrast, Python lists are managed by the Python interpreter, which introduces more overhead for slicing.

```python
#4.  How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a sample DataFrame
data = {
    'A': np.random.rand(10),
    'B': np.random.rand(10),
    'C': np.random.rand(10),
    'D': np.random.rand(10)
}

df = pd.DataFrame(data)

# Calculate the correlation matrix
corr_matrix = df.corr()

# Create a heatmap using Seaborn
plt.figure(figsize=(6, 4))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)

# Adding a title
plt.title('Correlation Matrix Heatmap')

# Display the heatmap
plt.show()
```

```python
#7. Write a program to perform element-wise multiplication of two NumPy arrays

import numpy as np

# Creating two NumPy arrays
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])

# Element-wise multiplication using *
result1 = arr1 * arr2

# Element-wise multiplication using numpy.multiply()
result2 = np.multiply(arr1, arr2)

# Display results
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Element-wise multiplication (using *):", result1)
print("Element-wise multiplication (using np.multiply()):", result2)
```

```python
#8. A Create a line plot with multiple lines using Matplotlib

import matplotlib.pyplot as plt
import numpy as np

# Sample data
x = np.arange(1, 11)
y1 = x * 2
y2 = x ** 2
y3 = np.log(x) * 10

# Creating the plot
plt.figure(figsize=(8, 5))
plt.plot(x, y1, marker='o', linestyle='-', color='b', label='y = 2x')
plt.plot(x, y2, marker='s', linestyle='--', color='r', label='y = x^2')
plt.plot(x, y3, marker='d', linestyle='-.', color='g', label='y = log(x)*10')

# Adding title and labels
plt.title("Multiple Line Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()  # Show legend
plt.grid(True)  # Show grid

# Display the plot
plt.show()
```

```python
#10. Create a histogram using Seaborn to visualize a distribution

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
data = np.random.normal(loc=50, scale=15, size=200)  # Normal distribution with mean=50, std=15

# Create the histogram
plt.figure(figsize=(8, 5))
sns.histplot(data, bins=20, kde=True, color='blue')

# Add labels and title
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram of Sample Data")

# Show the plot
plt.show()
```

```python
#11.  Perform matrix multiplication using NumPy

import numpy as np

# Define two matrices (2x3 and 3x2)
A = np.array([[1, 2, 3], [4, 5, 6]])  # 2x3 matrix
B = np.array([[7, 8], [9, 10], [11, 12]])  # 3x2 matrix

# Perform matrix multiplication
C = np.dot(A, B)  # Or you can use C = A @ B

# Display the result
print("Matrix A:")
print(A)
print("\nMatrix B:")
print(B)
print("\nResult of A * B:")
print(C)
```
